As discussed in the lectures, the cause of the auroral emissions at Earth are a consequence of disturbances to the Earth's magnetic environment, or magnetosphere. The magnetosphere is found to be in equilibrium when the dynamic pressure of the solar wind ($P_{SW}$) is balanced with the magnetic pressure produced by the Earth's magnetosphere, found at the magentopause ($P_{Mp}$): 

$$ P_{SW} = 2\rho_{SW}v^{2}_{SW} = \frac{B^{2}_{Mp}}{2\mu_{0}} = P_{Mp}$$ 

where $\rho_{SW}$ and $v_{SW}$ is the density and velocity of the solar wind respectively, $B_{Mp}$ is the dipolar magnetic field strength at the stand-off point (i.e., the tip of the magnetopause closest to the Sun) of the magnetopause and $\mu_{0}$ is the permeability of free space (a natural constant in S.I. units, as mentioned in Lecture 1). Although the plasma physics of this environment is beyond the scope of this course, we assume that the solar plasma particles are reflected at the magnetopause (i.e., bulk plasma behaviour cannot go through field lines), and this is shown by the factor of 2.

As we discussed in Lecture 2, changes in solar wind conditions can disturb the magnetosphere and change it from in an equlibrium state in two ways: $(1) P_{SW} > P_{MP}$: compression of magnetosphere and $(2) P_{SW} < P_{MP}$: expansion of the magnetosphere. As the change of pressure causes an inbalance, the stand-off point ($R_{Mp}$) will change location. This is given by the Chapman-Ferraro equation

$$R_{Mp} = R_{P}\left(\frac{B^{2}_{Mp}}{\rho_{SW}v^{2}_{SW}\mu_{0}}\right)^\frac{1}{6}$$

where $R_{P}$ is the planetary radius. For example, and increase in $P_{SW}$ from a higher $\rho_{SW}$ or $v_{SW}$ will reduce $R_{P}$, a classic signature of a compression. This kind of disturbance injects more energy into the magentosphere, via reconnection in the tail, causing stronger currents carrying particles from the outer regions to the Earth's poles - <b>this is the beginning of aurora production.</b>

How we measure this disturbance is either via spacecraft in the Earth's magnetosphere or on the ground measuring changes in the geomagnetic field and associated currents. Here, we will focus on the latter.


## Magnetometer Recap

As we dicussed in Lecture 2, we can measure the Earth's magnetic field strength using ground-based magnetometers. The most common magnetometer, both ground-based and on spacecraft, is the <i>fluxgate magnetometer</i>. Recall the basic operations of the fluxgate magnetometer:

<ol>
    <li> Wrap a coil of wire around a ferromganetic bar (i.e., a material that becomes magnetised when exposed to a magnetic field and remains magnetised when the field is removed).</li>
    <li> Run an electric current through the wire, magnetising the bar allowing it to generate it's own magnetic field - record the measurement </li>
    <li> Reversing the current, changes the field direction - record this measurement <b>(should be same magnitude as before, with the sign flipped)</b> </li>
    <li> Continuing to flip the current, both the magnetic field directions should cancel out. <b>HOWEVER</b> when an <i>external</i> field is present, it throws the induced field out of balance. </li>
    <li><b>Therefore, using this deviation we measure the external field i.e. the field of the Earth, recorded by the coils</b></li>
    <li> Commbining multiple fluxgate sensors allows us to measure the field in 3-dimensions.</li> 
</ol>

Therefore using these sensors, any disturbances of the external magnetic field in a given direction is easily detected and visulaised. Having a <b>network</b> of magnetometers, a known distance a part, allows you to measure the extent of the disturbance. Here we look at data from the <i>International Monitor for Auroral Geomagnetic Effects (IMAGE)</i> network, of which there are magnetometers dispersed at different latitudes in Finland.

## Improving auroral boundaries from Finnish magnetometer data

As discussed in L2, magnetometers measure the changes and disturbances of the magnetic field. This allows us to determine what is happening in the magnetosphere and how it affects the upper atmosphere. With a change in magnetic field, there is always a <b>change in current</b> that occurs in the ionosphere. The physics of which is far beyond this course, however using the IMAGE magenetometer network we can measure the <b>extent of the ionospheric current disturbances</b> and locate where the changes are. As shown in L2, current auroral models suffer drawbacks so this may be a potential solution on a local level.

This current in the upper atmosphere is known as the <b>electrojet current</b> and through modelling and mapping models, we can locate which magnetometers detect specific changes to this current. These are known as <b>auroral indicies</b> and provide us with an upper and lower bound. When the auroral oval is above a magnetometer, the current should be deflected (i.e., change direction) - and this is what we want to measure!

In the <i>.csv</i> file, <code>Walker2023_bound_filter_v2.csv</code> you have pre-processed data using these auroral indicies. Here we have determined the likley <b>auroral oval</b> from the indicies at each minute, using the strongest current in both directions. It is then assumed that the area in between the polar and equatorward points of the strongest current is the auroral oval during this time. The data spans 10-years of observations. The cell to read in the data is below, and has made into a dataframe <code>bound_df</code>.

The next data file contains solar wind parameters as measured by <a href= https://omniweb.gsfc.nasa.gov/form/dx1.html> OMNI spacecraft and ground stations </a>, which we can use to compare what we find from the Finnish magnetometers to solar wind conditions at the time.

The data in <code>bound_df</code>provided has the the following columns:
<ol><li><code>Date_UTC</code> - the date of the IMAGE data of the auroral oval</li>
    <li><code>MLT</code> - the associated MLT position of the oval</li>
    <li><code>P_bound</code>, <code>Eq_bound</code> - the polar and equatorward magnetic latitude of the oval at each time</li>
    <li><code>Oval_width</code> - the width of the auroral oval (i.e., the distance between <code>P_bound</code> and <code>Eq_bound</code>) </li>
    <li><code>Mag_in_MLT_range</code> - label determining whether any of the Finnish magnetometers lie within MLT range during this time of the observation (either <code>Yes</code> or <code>No</code>). Here, you can find the <a href="https://space.fmi.fi/image/www/?page=stations">Finnish IMAGE magnetometer coordinates</a> of the 11 stations</li> 
    <li> If <code>Yes</code>, the final column <code>Mag_in_lat_range</code> will then state how many magnetometers are within the latitude range. The idea being <b> the more magnetometers will mean that the auroral oval is larger, and likely mean a stronger geomagnetic storm</b>

The data in <code>omni_df</code>provided, solar wind and geomagentic values obtained/modelled from spacecraft and ground recievers, has the the following columns:
<ol><li><code>Date_UTC</code> - the date of the OMNI data product</li>
    <li><code>Bx, By, Bz</code> - the measured magnetic field strength in three components</li>
    <li><code>sw_density</code> - measured density of solar wind</li>
    <li><code>sw_pressure</code> - measured pressure of solar wind</li>
    <li><code>AE/AL/AU</code> - calcualted auroral indicies</li>


```python
import pandas as pd

bound_df = pd.read_csv('/home/yangz2/code/space_universe/data/A1/Walker2023_A1_model_data.csv')
bound_df['Date_UTC'] = pd.to_datetime(bound_df['Date_UTC'], format="%Y-%m-%d %H:%M:%S")

omni_df = pd.read_csv('/home/yangz2/code/space_universe/data/A1/OMNI_params_v2.txt', 
                         comment="#",sep=r"\s+",header=None, 
                      names=[ "Date_UTC", "Bx","By","Bz","sw_density","sw_pressure","AE","AL","AU"],
    parse_dates=[0])

# The above reads in the boundaries and solar wind data and creates a pandas dataframe

bound_df.set_index('Date_UTC', inplace=True)
omni_df.set_index('Date_UTC', inplace=True)
full_df = omni_df.join(bound_df, how='outer').reset_index()
full_df = full_df.dropna().reset_index(drop=True)
full_df = full_df.sort_values("Date_UTC").dropna().reset_index(drop=True)
#combine datasets to form full coverage of Earth's magnetosphere-ionsphere coupling

full_df.dropna() # remove all rows that have at least one missing value
```

Below is an example of the form of a <code>preprocessing</code> class. Here it is just a minimal identity preprocessor, i.e. no feature engineering. Use the below as a template if it helps! Also is added the format of how your model should be saved for testing.

```python
import numpy as np
import pandas as pd
import joblib

from sklearn.metrics import balanced_accuracy_score, accuracy_score
from sklearn.ensemble import HistGradientBoostingClassifier

# 导入预处理器类
from preprocessor_class import MyPreprocessor

# =========================
# 1) 目标：二分类
# =========================
TARGET = "Mag_in_MLT_range"


# =========================
# 2) 允许使用更多列：把"后面那些列"也作为特征输入
#    注意：不要把 TARGET 自己放进去
# =========================
# 这些列在你截图里明确存在；如果你的表里还有更多列，也可以继续加
EXTRA_FEATURES = ["MLT", "P_bound", "Eq_bound", "Oval_width", "Mag_in_lat_range"]

BASE_OMNI = ["Bx","By","Bz","sw_density","sw_pressure","AE","AL","AU"]

# FEATURES = 你要用的"原始输入列"
FEATURES = BASE_OMNI + EXTRA_FEATURES


# =========================
# 4) 生成 X/y，并处理 Yes/No 标签
# =========================
pre = MyPreprocessor(
    base_columns=FEATURES,
    lags_minutes=(5, 15, 30, 60),
    roll_minutes=(30, 60),
    add_physics_features=True
)
```

```python
X_all = pre.transform(full_df)
print("Target in Features?:", TARGET in X_all.columns)

# y 对齐：用 Date_UTC index 对齐
tmp = full_df.copy()
tmp["Date_UTC"] = pd.to_datetime(tmp["Date_UTC"])
tmp = tmp.sort_values("Date_UTC").set_index("Date_UTC")

y_raw = tmp.loc[X_all.index, TARGET]

# Yes/No → 1/0
y_str = y_raw.astype(str).str.strip().str.lower()
mapping = {"yes": 1, "no": 0}
y_all = y_str.map(mapping)

# 丢掉异常值并同步 X
mask = y_all.notna()
X_all = X_all.loc[mask]
y_all = y_all.loc[mask].astype(int)

print("After feature engineering:", X_all.shape, y_all.shape)
print("Positive ratio:", float(y_all.mean()))


# =========================
# 5) 时间切分（按顺序 70/15/15）
# =========================
n = len(X_all)
i_train = int(0.70 * n)
i_val = int(0.85 * n)

X_train, y_train = X_all.iloc[:i_train], y_all.iloc[:i_train]
X_val, y_val     = X_all.iloc[i_train:i_val], y_all.iloc[i_train:i_val]
X_test, y_test   = X_all.iloc[i_val:], y_all.iloc[i_val:]

print("Splits:", len(X_train), len(X_val), len(X_test))


# =========================
# 6) 训练：加权处理类不平衡（即使你加了边界特征，仍建议保留）
# =========================
clf = HistGradientBoostingClassifier(
    learning_rate=0.08,
    max_depth=8,
    max_iter=400,
    random_state=42
)

pos = int(y_train.sum())
neg = int(len(y_train) - pos)
pos_weight = (neg / max(pos, 1))
w_train = np.where(y_train.values == 1, pos_weight, 1.0)
print(f"Train pos={pos}, neg={neg}, pos_weight≈{pos_weight:.1f}")

clf.fit(X_train, y_train, sample_weight=w_train)


# =========================
# 7) 评估：用验证集选阈值（maximize balanced acc）
# =========================
def rmse_prob(y_true, y_prob):
    y_true = np.asarray(y_true, dtype=float)
    y_prob = np.asarray(y_prob, dtype=float)
    return float(np.sqrt(np.mean((y_prob - y_true) ** 2)))

def find_best_threshold_balacc(y_true, prob):
    # 兼容极端概率分布：从很小阈值扫到 0.8
    ts = np.concatenate([np.logspace(-6, -2, 60), np.linspace(0.01, 0.80, 80)])
    best_t, best_bacc = 0.5, -1.0
    for t in ts:
        pred = (prob >= t).astype(int)
        bacc = balanced_accuracy_score(y_true, pred)
        if bacc > best_bacc:
            best_bacc = bacc
            best_t = float(t)
    return best_t, float(best_bacc)

prob_train = clf.predict_proba(X_train)[:, 1]
prob_val   = clf.predict_proba(X_val)[:, 1]
prob_test  = clf.predict_proba(X_test)[:, 1]

# 概率分布看看（判断模型是否开始输出正类）
for name, p in [("TRAIN", prob_train), ("VAL", prob_val), ("TEST", prob_test)]:
    qs = np.quantile(p, [0.5, 0.9, 0.99, 0.999])
    print(f"{name} prob quantiles [50%,90%,99%,99.9%] = {qs}")

best_t, best_val_bacc = find_best_threshold_balacc(y_val, prob_val)
print(f"Best threshold on VAL (by balanced acc): t={best_t:.6f}, val_bacc={best_val_bacc:.4f}")

def eval_with_threshold(name, y, prob, t):
    pred = (prob >= t).astype(int)
    rmse = rmse_prob(y, prob)
    bacc = balanced_accuracy_score(y, pred)
    acc  = accuracy_score(y, pred)
    pos_rate = float(pred.mean())
    print(f"[{name}] RMSE(prob): {rmse:.4f} | Acc: {acc:.4f} | Balanced Acc: {bacc:.4f} | pred_pos_rate: {pos_rate:.4f} | thr={t:.6f}")

eval_with_threshold("TRAIN", y_train, prob_train, best_t)
eval_with_threshold("VAL  ", y_val,   prob_val,   best_t)
eval_with_threshold("TEST ", y_test,  prob_test,  best_t)


# =========================
# 8) 保存提交文件
# =========================
bundle = {
    "target": TARGET,
    "model": clf,
    "preprocessor": pre,   # 预处理器里包含所有特征工程
}
joblib.dump(bundle, "student_model.joblib")
print("Saved student_model.joblib")
```
